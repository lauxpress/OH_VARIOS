Buscando datasets para utilizar en el trabajo recorde que para un proyecto habia encontrado data de la profundidad del rio parana medido en el puerto de Corrientes (mi ciudad) y pense que seria un buen parametro para analizar y entender.

```python
import pandas as pd
import plotly.express as px
# Load data from CSV file
df = pd.read_csv('/content/drive/MyDrive/Series Temporales/Altura_Parana.csv')

# Convert 'Fecha' column to datetime format
df['Fecha'] = pd.to_datetime(df['Fecha'])

df.head(10)
```

Los datos se encuentran en la pagina oficial de prefectura y cuentan con dos lecturas diarias de profundidad y se puede apreciar la ultima crecida grande del Rio en los ultimos meses.

```python
fig = px.line(df, x='Fecha', y='Registro (m)', title='Profundidad del Parana')
fig.show()
```

Sin embargo creo que no aporta la suficiente cantidad de datos en cuanto al tiempo, ya que lo que se habia dicho en clases es que debia contar con mas de 2 aos de data para encontrar caracteristicas de seasonality. En esta grafica notamos que el periodo observado es de hecho muy corto y que la variacion tan fuera de lo comun como es la subida inmensa que presento el rio en noviembre no puede ser generalizada y no podra ser predecida de manera correcta la periodicidad ni el comportamiento futuro del rio.
